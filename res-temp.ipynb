```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dfH = pd.read_csv('res-temp_hierro.csv')
dfH.head()
```

```python
dfC = pd.read_csv('res-temp_cobre.csv')
dfC.head()
```

```python
x,y = dfH['temperatura'],dfH['resistencia']

fig, ax = plt.subplots(figsize=(8, 8))

ax.scatter(x,y)
ax.set_title('Temperatura vs resistencia (Hierro)')
ax.set_ylabel('Resistencia (Ω)')
ax.set_xlabel('Temperatura (°c)')
```

```python
x,y = dfC['temperatura'],dfC['resistencia']

coef = np.polyfit(x, y, 1)  # Grado 1 para regresión lineal
poly1d_fn = np.poly1d(coef)

fig, ax = plt.subplots(figsize=(8, 8))

ax.scatter(x,y)
ax.plot(x, poly1d_fn(x),c='r',label='Pendiente ajustada = 0.029Ω/c°')

ax.set_title('Temperatura vs resistencia (Cobre)')
ax.set_ylabel('Resistencia (Ω)')
ax.set_xlabel('Temperatura (°c)')
ax.legend()
```
